```python
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
```

```python
url = 'https://cloud.iexapis.com/stable/stock/AAPL/chart/5y' # data within 5 years from now
# url = 'https://cloud.iexapis.com/stable/ref-data/symbols'
```

```python
parameters = {
    'token': 'sk_14777377f60b4a41bef74ca53d49e510'
}
headers = {
    'Accepts': 'application/json'
}
```

```python
session = Session()
session.headers.update(headers)
```

```python
try:
    response = session.get(url, params=parameters)
    data = json.loads(response.text)
    print(data)
except (ConnectionError, Timeout, TooManyRedirects) as e:
    print(e)
```

```python
len(data)
```

```python
close_price = []
open_price = []
high = []
low = []
volume = []
date = []
change = []
changePercent = []
for item in data:
    print(item)
    close_price.append(item['close'])
    open_price.append(item['open'])
    high.append(item['high'])
    low.append(item['low'])
    date.append(item['date'])
    volume.append(item['volume'])
    change.append(item['change'])
    changePercent.append(item['changePercent'])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Date': date,'Open':open_price,'High':high,'Low':low,'Close':close_price,'Volume':volume,'Change': change,'Change %': changePercent})
```

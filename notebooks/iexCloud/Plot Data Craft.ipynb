```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pytz
```

```python
asks = pd.read_csv('14122020_asks.csv')
bids = pd.read_csv('14122020_bids.csv')
trades = pd.read_csv('14122020_trades.csv')
```

```python
est = pytz.timezone('US/Eastern')
bids['timestamp'] = pd.to_datetime(bids['timestamp'], unit='ms')
asks['timestamp'] = pd.to_datetime(asks['timestamp'], unit='ms')
trades['timestamp'] = pd.to_datetime(trades['timestamp'], unit='ms')
# trades['timestamp'] = trades['timestamp'].dt.tz_localize(pytz.utc)
# trades['timestamp'] = trades['timestamp'].dt.tz_convert(est)
```

```python
bids.sort_values('timestamp', ascending=True)
asks.sort_values('timestamp', ascending=True)
trades.sort_values('timestamp', ascending=True)
```

```python
open_price = 122.6
high_price = 123.35
low_price = 121.54
close_price = 121.78
```

```python
fig = plt.figure(num=3, figsize=(20,10))
plt.plot(trades['timestamp'], trades['price'], 'ob', label='Trades')
plt.plot(bids['timestamp'], bids['price'], 'og', label='Bids')
plt.plot(asks['timestamp'], asks['price'], 'or', label='Asks')
plt.axhline(open_price, linestyle='-', label='Open Price')
plt.axhline(high_price, linestyle=':', label='High Price')
plt.axhline(low_price, linestyle='-.', label='Low Price')
plt.axhline(close_price, linestyle='--', label='Close Price')
plt.ylim(low_price - 0.2, high_price + 0.2)
plt.legend()
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Bids, Asks and Trade with AAPL Stock on 14/12/2020')
plt.show()
```

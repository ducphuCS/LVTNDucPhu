```python
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
```

```python
url = 'https://poloniex.com/public?'
```

```python
parameters = {
    'command': 'returnChartData',
    'currencyPair': 'BTC_ETH',
    'start': 1448928000,
    'end': 1606780800,
    'period': 14400 # candlestick period in second
}
headers = {
    'Accepts': 'application/json'
}
```

```python
session = Session()
session.headers.update(headers)
```

```python
try:
  response = session.get(url, params=parameters)
  data = json.loads(response.text)
  print(data)
except (ConnectionError, Timeout, TooManyRedirects) as e:
  print(e)
```

```python
dates = []
# highest price within this candle
high = []
# lowest price within this candle
low = []
# price at the start and the end of this candle
open_price = []
close_price = []
# total volume transacted within this candle
volumes = []
# total amount of base transacted within this candle
quote_volume = []
# average price within this candle
price = []
```

```python
import pandas as pd
```

```python
for item in data:
    dates.append(pd.to_datetime(item['date'],unit='s'))
    high.append(item['high'])
    low.append(item['low'])
    open_price.append(item['open'])
    close_price.append(item['close'])
    volumes.append(item['volume'])
    quote_volume.append(item['quoteVolume'])
    price.append(item['weightedAverage'])
print(dates)
```

```python
df = pd.DataFrame({'Dates': dates,'High':high,'Low':low,'Open':open_price,'Close':close_price,'Volume':volumes,'Quote Volume':quote_volume,'Average Price':price})
```

```python
import json
import pandas as pd
```

```python
data = pd.read_json('coinmarketcap.json')
data
```

```python
with open('coinmarketcap.json', 'r') as f:
    data_json = json.load(f)
```

```python
data_json
```

```python
# select name, price, volume, last updated time ... for now
price = []
name = []
volume = []
last_updated = []
for item in data_json:
    name.append(item['name'])
    volume.append(item['quote']['USD']['volume_24h'])
    price.append(item['quote']['USD']['price'])
    last_updated.append(item['last_updated'])
print(price)
```

```python
name = pd.Series(name)
price = pd.Series(price)
volume = pd.Series(volume)
last_updated = pd.Series(last_updated)
```

```python
df = pd.DataFrame({'Name': name, 'Price': price, 'Volume': volume, 'Last updated': last_updated})
print(df)
```

```python
df.to_csv('coinmarketcap.csv', index=False)
```
